```python
# to create a model
from sklearn import tree
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score

#to separate independent variable from iris dataset
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']

#splitting the data into training and testing 
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,y, test_size = 0.30, random_state = 42)

#creating decision tree classifier and train it with training dataset
Iris_clf = DecisionTreeClassifier(criterion = 'gini', min_samples_split=2)
Iris_clf.fit(Xtrain, Ytrain)


#plot the tree
tree.plot_tree(Iris_clf)
```

```python
# to plot the decision tree using graphviz library

dot_data = tree.export_graphviz(Iris_clf, out_file=None)

graph = graphviz.Source(dot_data)
graph
```
